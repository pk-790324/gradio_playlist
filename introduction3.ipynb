```python
import ollama
```

```python
def chatbot(text):
    response=ollama.generate(model='tinyllama:1.1b',prompt='text')
    return response['response']
    
```

```python
demo=gr.Interface(
    fn=chatbot,
    inputs=['text'],
    outputs=['text']
)
demo.launch()
```
